```python
import numpy as np
import pandas as pd
import math
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('Creditcard_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

```python
x=df.drop(['Class'],axis=1)
y=df['Class']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2)
```

```python
#solving the problem using overfitting
rm=RandomOverSampler(random_state=42)
x_train,y_train=rm.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
df=pd.DataFrame(x_train)
df['Class']=y_train
df
```

```python
#SAMPLING
```

```python
s=[]
```

```python
#Random sampling
random_sample_size=math.ceil((1.96**2 * 0.5 * (1 - 0.5)) / (0.05**2))
s1=df.sample(random_sample_size)
s1['Class'].value_counts()
s.append(s1)
```

```python
#Stratified sampling
stratified_sample_size=math.ceil((1.96**2 * 0.5 * (1 - 0.5)) / ((0.05/0.7)**2))
s2=df.groupby('Class',group_keys=False).apply(lambda x: x.sample(stratified_sample_size))
s2['Class'].value_counts()
s.append(s2)
```

```python
#Systematic sampling
systematic_sample_size=math.ceil((1.96**2 * 0.5 * (1 - 0.5)) / (0.05**2))
k=math.ceil(len(df)/systematic_sample_size)
s3=df.iloc[::k]
s3['Class'].value_counts()
s.append(s3)
```

```python
#Bootstrap sampling
bootstrap_sample_size=math.ceil((1.96**2 * 0.5 * (1 - 0.5)) / (0.05**2))
s4=df.sample(bootstrap_sample_size,replace=True)
s4['Class'].value_counts()
s.append(s4)
```

```python
#cluster sampling
cluster_sample_size=math.ceil((1.96**2 * 0.5 * (1 - 0.5)) / ((0.05/1.5)**2))
K=math.ceil(len(df)/cluster_sample_size)
clusters=2
data=None
for k in range(K):
    sample_k = df.sample(cluster_sample_size)
    sample_k["cluster"] = np.repeat(k,len(sample_k))
    df_new = df.drop(index = sample_k.index)
    data = pd.concat([data,sample_k],axis = 0)

random_chosen_clusters = np.random.randint(0,K,size = clusters)
s5 = data[data.cluster.isin(random_chosen_clusters)]
s5.drop(['cluster'], axis=1, inplace=True)
s5['Class'].value_counts()
s.append(s5)
```

```python
mat=pd.DataFrame(columns=['Simple-Random','Stratified','Systematic','Bootstrap','Cluster'], index=['Random Forest','SVM','Logistic Regression','XGBoost','KNN'])
```

```python
for i in range(0,5):
  j=0
  inp=s[i].drop(['Class'],axis=1)
  out=s[i]['Class']
  inp_train, inp_test, out_train, out_test = train_test_split(inp ,out ,test_size=0.25, shuffle=True)

  # Applying Random Forest Classifier
  rfc = RandomForestClassifier()
  rfc.fit(inp_train, out_train)
  y_pred = rfc.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  mat.iloc[j,i]=acc*100


  # Applying SVM
  svm = SVC(kernel='rbf')
  svm.fit(inp_train, out_train)
  y_pred=svm.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  mat.iloc[j+1,i]=acc*100


  # Applying Logistic Regression
  lr = LogisticRegression(max_iter=2000)
  lr.fit(inp_train, out_train)
  y_pred = lr.predict(inp_test)
  acc = accuracy_score(out_test, y_pred)
  mat.iloc[j+2,i]=acc*100


  # Applying XGBoost Classifier
  xgb = XGBClassifier().fit(inp_train, out_train)
  y_predict = xgb.predict(inp_train)
  acc = accuracy_score(y_predict, out_train)
  mat.iloc[j+3,i]=acc*100

  # Applying knn
  scaler=StandardScaler()
  inp_train_scaled=scaler.fit_transform(inp_train)
  inp_test_scaled=scaler.fit_transform(inp_test)

  knn=KNeighborsClassifier()
  knn.fit(inp_train_scaled,out_train)

  out_pred=knn.predict(inp_test)
  acc=accuracy_score(out_test,out_pred)
  mat.iloc[j+4,i]=acc*100

```

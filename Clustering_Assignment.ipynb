```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
data = get_data("jewellery")
```

```python
data.to_csv("data.csv")
from google.colab import files
```

**1 K-MEANS**

**1.1**

```python
from pycaret.clustering import *
setup(data,verbose =False)

print("3 clusters ")
kmeansModel1 = create_model('kmeans',num_clusters=3)

print("4 clusters ")
kmeansModel2 = create_model('kmeans',num_clusters=4)

print("5 clusters")
kmeansModel3 = create_model('kmeans',num_clusters=5)
```

```python
plot_model(kmeansModel3, plot='elbow')
```

```python
plot_model(kmeansModel1, plot='cluster')
```

```python
plot_model(kmeansModel1, plot='tsne')
```

```python
plot_model(kmeansModel1, plot= 'silhouette')
```

```python
plot_model(kmeansModel1, plot = 'distribution')
```

```python
setup(data = data, normalize = True, normalize_method = 'zscore', verbose= False)

print("3 clusters ")
kmeansModel1 = create_model('kmeans',num_clusters=3)

print("4 clusters ")
kmeansModel2 = create_model('kmeans',num_clusters=4)

print("5 clusters")
kmeansModel3 = create_model('kmeans',num_clusters=5)
```
